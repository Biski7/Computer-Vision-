```python
# Resizing images
plt.imshow(img_1)
```

```python
img_1.shape
```

```python
img_resized = cv2.resize(img_1, (800,640) )
plt.imshow(img_resized)
```

```python
# resizing by ratio
width_ratio = 0.5 #50% of original
height_ratio = 0.3

img_resized_ratio = cv2.resize(img_1, (0,0), img_1, width_ratio, height_ratio)
plt.imshow(img_resized_ratio)
```

```python
# flipping images
img_flip = cv2.flip(img_1, 0)
plt.imshow(img_flip)
```

```python
img_flip = cv2.flip(img_1, 1)
plt.imshow(img_flip)
```

```python
img_flip = cv2.flip(img_1, -1)
plt.imshow(img_flip)
```

```python
# Writing to new file
cv2.imwrite('modified_image.jpg', img_1)
```
